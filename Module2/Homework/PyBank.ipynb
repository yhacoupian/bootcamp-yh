```python
import pandas as pd
import numpy as np
```

```python
p_and_l_df = pd.read_csv("02-Homework_02-Python_Instructions_PyBank_Resources_budget_data.csv")
```

```python
p_and_l_df.head()
```

```python
# Number of months
len(p_and_l_df.index)
```

```python
# shift the value in each column one place down 
p_and_l_df['lag'] = p_and_l_df['Profit/Losses'].shift(1)
```

```python
p_and_l_df.head()
```

```python
# replace NaN with zero inplace 
p_and_l_df['lag'].fillna(0, inplace=True)
```

```python
p_and_l_df['lag'] = p_and_l_df['lag'].astype('int32')
```

```python
p_and_l_df.head()
```

```python
p_and_l_df['profit increase/decrease'] = p_and_l_df['Profit/Losses'] - p_and_l_df['lag']
```

```python
p_and_l_df.head()
```

```python
p_and_l_df.loc[0, 'profit increase/decrease'] = 0
```

```python
p_and_l_df
```

```python
total_months = len(p_and_l_df.index)
total = p_and_l_df['Profit/Losses'].sum()
average = p_and_l_df['profit increase/decrease'].sum()/(len(p_and_l_df.index)-1) # can not use the mean 
maximum = p_and_l_df['profit increase/decrease'].max()
minimum = p_and_l_df['profit increase/decrease'].min()
```

```python
print(total_months, total, average, maximum, minimum)
```

```python
# deduct 1 from the overall length to discard the first 
print(p_and_l_df['profit increase/decrease'].sum()/(len(p_and_l_df.index)-1))
```
